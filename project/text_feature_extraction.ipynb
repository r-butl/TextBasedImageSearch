### Using a simple model we can use a Tokenizer
- For this first section we are looking at a single sentence and the extraction from that:

---

````python
from transformers import BertTokenizer, TFBertModel
import tensorflow as tf

#load pre-trained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = TFBertModel.from_pretrained("bert-base-uncased")

#sample sentence
text = "Natural language processing with transformers is powerful."

#tokenize and encode return tensorflow tensors
inputs = tokenizer(text, return_tensors='tf')

#get embeddings
outputs = model(**inputs)

#extract the last hidden state
last_hidden_state = outputs.last_hidden_state

#CLS token embedding
cls_embedding = last_hidden_state[:, 0, :]  # shape: (1, hidden_size)
print("CLS token embedding shape:", cls_embedding.shape)
print("First 5 values of sentence embedding:", cls_embedding[0][:5])

#convert to numpy for easier handling
cls_vector = cls_embedding.numpy()[0]

#print first 10 values 
print("First 10 features of the CLS embedding:")
print(cls_vector[:10])
````

# Moving forward we look at the most significant Features within the CLS Embedding 

````python
import numpy as np
import matplotlib.pyplot as plt

#sort by absolute value and get top 20 most significant features
top_k = 20
indices = np.argsort(np.abs(cls_vector))[::-1][:top_k]
top_values = cls_vector[indices]

plt.figure(figsize=(12, 5))
plt.bar(range(top_k), top_values)
plt.xticks(range(top_k), labels=[f'feat_{i}' for i in indices], rotation=45)
plt.title("Top 20 Most Significant Features in CLS Embedding")
plt.ylabel("Feature Value")
plt.tight_layout()
plt.show()

#convert to numpy for easier handling
cls_vector = cls_embedding.numpy()[0]

#print first 10 values
print("First 10 features of the CLS embedding:")
print(cls_vector[:10])
````

# Next we will look at clustering those sentences based upon the content

````python
#simple sentences
sentences = [
    "A cat is sitting on a couch.",
    "Two dogs are playing in the park.",
    "A man riding a bicycle on the road.",
    "A woman is cooking in the kitchen.",
    "Children are playing soccer on a field.",
    "A plate of sushi on a wooden table.",
    "A dog chasing a ball in the backyard.",
    "A person skiing down a snowy hill.",
    "A chef preparing food in a restaurant.",
    "Kids running through a sprinkler in summer."
]


#get CLS embeddings
def get_cls_embedding(text):
    inputs = tokenizer(text, return_tensors='tf', truncation=True, padding=True)
    outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].numpy()[0]

embeddings = np.array([get_cls_embedding(sent) for sent in sentences])
print("Embeddings shape:", embeddings.shape)  
````

# Next we will define some of the clustering of the sentences based upon the embeddings

````python
from sklearn.cluster import KMeans

#choose number of clusters
k = 6
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(embeddings)

#show clustering
for sent, label in zip(sentences, clusters):
    print(f"[Cluster {label}] {sent}")

````

# Next we will diplay those Clusterings

````python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced = pca.fit_transform(embeddings)

plt.figure(figsize=(8, 6))
for i in range(len(sentences)):
    plt.scatter(reduced[i, 0], reduced[i, 1], c=f"C{clusters[i]}", label=f"Cluster {clusters[i]}")
    plt.text(reduced[i, 0]+0.01, reduced[i, 1]+0.01, sentences[i], fontsize=9)

plt.title("Sentence Clustering (PCA Projection)")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````

# In the last section we will consider a better model for clustering:
---
### This model is the SBERT model
- Great for sentence comparison
- Trained specifically for sentence level tasks
- Uses the BERT model with pooling and triplet architecture

````python
import tensorflow_hub as hub

# Load a pre-trained sentence encoder
sbert_model = hub.load("https://tfhub.dev/google/universal-sentence-encoder/4")

# Embed your captions
embeddings = sbert_model(sentences)  # shape: (N, 512)


kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(embeddings)

# Show cluster assignment
for sent, label in zip(sentences, clusters):
    print(f"[Cluster {label}] {sent}")

pca = PCA(n_components=2)
reduced = pca.fit_transform(embeddings)

plt.figure(figsize=(8, 6))
for i in range(len(sentences)):
    plt.scatter(reduced[i, 0], reduced[i, 1], c=f"C{clusters[i]}", label=f"Cluster {clusters[i]}")
    plt.text(reduced[i, 0]+0.01, reduced[i, 1]+0.01, sentences[i], fontsize=9)

plt.title("Sentence Clustering (PCA Projection)")
plt.xlabel("PC 1")
plt.ylabel("PC 2")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
````
